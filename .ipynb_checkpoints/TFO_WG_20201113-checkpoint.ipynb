Notes on wave glider operation for 2020 TFO. 
-------


Operational notes (from Nick Statom)
-------
* I did not see a 'fixed heading' command in the command archive and the target waypoint in the telemetry still said '12' but the vehicle still was not on the correct course when I took over. I resent the Follow Sequential Command and it seemed to snap out of it. I am guessing this behavior was due to the GPS outages that happened around 4:30 but just something to keep an eye on

Laurent probably has more checks but I listed a few

* check navigation
--if ship traffic then move via fixed heading (Follow Fixed Heading button on home page) or create new waypoints (More Commands > Command Type > New Waypoint to send each waypoint then use the Follow Sequential Course command on the home page)
--if gps outage then move via fixed heading, use TRACKER_SV2 for position updates, and put back on sequential course if needed

* check telemetry (vehicle icon > telemetry)
--check past and present ground speed to get an idea of the currents (you can also click on each past telemetry point on the map as well to check speed, heading and distance traveled)
--check past and present power to get an idea of power draw (you can also hover over top right battery indicator or click on each map telemetry point as well), go to vehicle icon > power ctl statuses for more detailed power overview showing individual payload power and battery health

* check payload data for problematic behavior or to influence a new navigation/sampling strategy via luc script (run quick_view_Kelvin.m with all dependent m files once connected to the VPN and able to access waveglider2.ucsd.edu computer) 

* check to make sure pressure/humidity of payloads haven't changed (Show System Menu > Data > Payload Status)

* check vehicle health (click on colored circle indicator on top of home page to view past/present alarms and click status link to clear)

* turn on/off light during at night/day (click light button on home page, green = on, gray = off)

Other FAQ (answered by Laurent)
----
> Q: 1) I cannot see the way points when I am on Kelvin's interface (but
> can see them if I go to a different WG's interface).
* you need to press the toggle button which is in the map utility panel
(see png attached)

>
> Q: 2) I do not see any boat traffic, even though I have selected the
> usual AIS settings.
* Try to zoom out to look for larger field of field + select a longer
history (so you can see ships that were detected earlier)
